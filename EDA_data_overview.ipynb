```python
!pip install --upgrade pandas pyarrow fastparquet

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

print("\nĐang đọc dữ liệu...")

# 2. Dùng tuyệt chiêu thử engine: Nếu pyarrow lỗi, tự động chuyển sang fastparquet
try:
    users = pd.read_parquet('data/processed/users.parquet', engine='pyarrow')
    movies = pd.read_parquet('data/processed/movies.parquet', engine='pyarrow')
    ratings = pd.read_parquet('data/processed/ratings.parquet', engine='pyarrow')
    print("✅ Đọc file thành công bằng pyarrow!")
except:
    print("pyarrow vẫn kén file, tự động chuyển sang fastparquet...")
    users = pd.read_parquet('data/processed/users.parquet', engine='fastparquet')
    movies = pd.read_parquet('data/processed/movies.parquet', engine='fastparquet')
    ratings = pd.read_parquet('data/processed/ratings.parquet', engine='fastparquet')
    print("✅ Đọc file thành công bằng fastparquet!")

print("\n=== BỨC TRANH TOÀN CẢNH ===")
print(f"Số lượng User: {users.shape[0]:,}")
print(f"Số lượng Phim: {movies.shape[0]:,}")
print(f"Số lượng Đánh giá (Ratings): {ratings.shape[0]:,}\n")

print("=== DỮ LIỆU PHIM (CẬP NHẬT MỚI) ===")
display(movies.head())
```
