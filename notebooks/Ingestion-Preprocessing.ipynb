```python
import pandas as pd
```

```python
users = pd.read_csv(
    "users.dat",
    sep="::",
    engine="python",
    names=["UserID","Gender","Age","Occupation","ZipCode"]
)

users.head()
```

```python
# Load the Movies dataset
# The encoding is set to 'latin-1' to correctly read special characters in movie titles.

movies = pd.read_csv(
    "movies.dat",
    sep="::",
    engine="python",
    encoding="latin-1",
    names=["MovieID","Title","Genres"]
)

movies.head()
```

```python
ratings = pd.read_csv(
    "ratings.dat",
    sep="::",
    engine="python",
    names=["UserID","MovieID","Rating","Timestamp"]
)

ratings.head()
```

```python
ratings["Timestamp"] = pd.to_datetime(ratings["Timestamp"], unit="s")

ratings.head()
```

```python
users["Gender"] = users["Gender"].map({
    "F":0,
    "M":1
})

users.head()
```

```python
movies["Genres"] = movies["Genres"].str.split("|")

movies.head()
```

```python
users.to_parquet("users.parquet", index=False)
movies.to_parquet("movies.parquet", index=False)
ratings.to_parquet("ratings.parquet", index=False)
```

```python
users_parquet = pd.read_parquet("users.parquet")
movies_parquet = pd.read_parquet("movies.parquet")
ratings_parquet = pd.read_parquet("ratings.parquet")

users_parquet.head()
```
